```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
userage=np.array([12,23,5,2,45,23])
sallary=np.array([23,52,45,12,33,43])
```

```python
plt.plot(userage,sallary,marker=".",ms=10,mec="green",mfc="red",linestyle="dashdot",color="yellow")
```

```python
plt.grid()
plt.xlabel("amout")
plt.ylabel("number")
plt.title("my data")
plt.plot(userage,sallary,marker=".",ms=10,mec="green",mfc="red",linestyle="dashdot",color="yellow")

# plt.annotate(xy=(40,40),text="hello")

for i in range(len(userage)):
    plt.annotate(xy=(userage[i],sallary[i]),text =f"({userage[i]},{sallary[i]})")
```

```python
month = np.array([2,4,5,6,7,3,8,2,5,6])
sales=np.array([200,400,232,543,343,213,566,657,234,123])
prfot=[200,300,400,500,600,700,800,899,100,433]

colorr=["red","green","orange","black","blue","pink","gray","lightgreen","purple","red"]
plt.grid()
plt.scatter(month,sales,sizes =prfot,color=colorr,label="data")
for i in range(len(month)):
    plt.annotate(xy=(month[i],sales[i]),text =f"({month[i]},{sales[i]})")
plt.legend()
```
