```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([2,3,4,5,6])
x
```

```python
y=np.array([5,6,9,8,2])
```

```python
plt.plot(x,y)
```

```python
# #marker color
# ms = marker size
# mec-->marker edge color
# mfc--->marker face color
plt.plot(x,y,marker="*",ms = 20,mec= "red" ,mfc="yellow",color="black",linewidth=2,linestyle="--")
```

```python
plt.plot(x,y,marker="*",ms = 20,mec= "red" ,mfc="yellow",color="black",linewidth=2,linestyle=":")
```

```python
months=np.array([1,2,3,4,5,6,7])
profit=([20,10,15,45,34,23,20])
plt.plot(months,profit)
```

```python
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This year revenue",color="red")
plt.plot(months,profit,marker=".",ms=10)
```

```python
#aply grid
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This year revenue",color="red")
plt.plot(months,profit,marker=".",ms=10)
plt.grid()
```

```python
#aply grid
plt.xlabel("All Months")
plt.ylabel("My Profit Per Month")
plt.title("This year revenue",color="red")
plt.plot(months,profit,marker=".",ms=10)
plt.grid(linestyle="-.",color="pink")
```
